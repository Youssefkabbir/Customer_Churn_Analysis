# Importing Necessary Libraries


```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
```

#  Load Dataset


```python

# Load dataset
df = pd.read_csv(r"C:\Users\kabir\Desktop\Projects\Business_Analyst\Customer_Churn_Analysis\Python\data\raw\Bank Customer Churn Prediction.csv")
```

#  Data Cleaning & Standardization


```python
# Standardize column names
df.columns = df.columns.str.strip().str.lower()
```

# Display Basic Dataset Information


```python
# Display basic dataset information
print("Dataset Overview:")
print(df.info())
```

```python
print("\nSummary Statistics:")
print(df.describe())
```

```python
print("\nMissing Values:")
print(df.isnull().sum())
```

#  Correlation Heatmap


```python
# Correlation heatmap
numeric_df = df.select_dtypes(include=['number'])
plt.figure(figsize=(10,6))
sns.heatmap(numeric_df.corr(), annot=True, cmap='coolwarm', fmt='.2f')
plt.title("Feature Correlation Heatmap")
plt.show()
```

#  Age Distribution


```python
# Age Distribution
if 'age' in df.columns:
    plt.figure(figsize=(8,5))
    sns.histplot(df['age'], bins=20, kde=True, color='blue')
    plt.title("Age Distribution")
    plt.xlabel("Age")
    plt.ylabel("Frequency")
    plt.show()
```

# Balance Distribution

```python
# Balance Distribution
if 'balance' in df.columns:
    plt.figure(figsize=(8,5))
    sns.histplot(df['balance'], bins=20, kde=True, color='green')
    plt.title("Balance Distribution")
    plt.xlabel("Balance")
    plt.ylabel("Frequency")
    plt.show()
```

# Churn Rate by Gender

```python
# Churn Rate by Gender
if 'gender' in df.columns and 'churn' in df.columns:
    plt.figure(figsize=(6,4))
    sns.countplot(data=df, x='gender', hue='churn', palette='coolwarm')
    plt.title("Churn Rate by Gender")
    plt.show()
```

# Churn Rate by Country

```python
# Churn Rate by Country
if 'country' in df.columns and 'churn' in df.columns:
    plt.figure(figsize=(8,5))
    sns.countplot(data=df, x='country', hue='churn', palette='Set2')
    plt.title("Churn Rate by Country")
    plt.show()
```

# Churn Rate by Credit Score

```python
# Churn Rate by Credit Score
if 'credit_score' in df.columns and 'churn' in df.columns:
    plt.figure(figsize=(10,6))
    sns.boxplot(data=df, x='churn', y='credit_score', hue='churn', palette='viridis', legend=False)
    plt.title("Credit Score Distribution by Churn Status")
    plt.show()
```
